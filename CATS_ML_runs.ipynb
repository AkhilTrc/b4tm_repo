# Support Vector Classification

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Separate the features and labels
#
X = trainset.iloc[:, :-1].values
y = trainset.iloc[:, -1].values

# Split the dataset into training and testing sets
#
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Scale the features
#
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Train the SVM model
#
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
#
y_pred = classifier.predict(X_test)

# Evaluate the performance of the model
#
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
```

# Naive Bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

# CV

```python
from sklearn.model_selection import LeaveOneOut, KFold
from sklearn.svm import SVC
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
dataset = pd.read_csv('Data/Train_call.txt', sep='\t')
labels = pd.read_csv('Data/Train_clinical.txt', sep='\t')

dataset_transposed = dataset.T
dataset_transposed = dataset_transposed.drop(['Start', 'End', 'Nclone', 'Chromosome'])
dataset_transposed = dataset_transposed.reset_index()

trainset = pd.concat([dataset_transposed, labels], axis=1)
trainset = trainset.set_index('Sample')
trainset = trainset.drop(columns = ['index'])
```

```python
# Separate the features and labels
#
X = trainset.iloc[:, :-1].values
y = trainset.iloc[:, -1].values

# Split the dataset into training and testing sets
#
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
loo = LeaveOneOut()
loo_scores = []
```

```python
for train_idx, test_idx in loo.split(X):
    # split the data into train and test sets for the outer loop
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    # define the inner loop
    inner_cv = KFold(n_splits=10)
    svc = SVC(C = 0.1, kernel='linear')
    inner_scores = []
    
    for inner_train_idx, inner_test_idx in inner_cv.split(X_train):
        # split the data into train and test sets for the inner loop
        X_inner_train, X_inner_test = X_train[inner_train_idx], X_train[inner_test_idx]
        y_inner_train, y_inner_test = y_train[inner_train_idx], y_train[inner_test_idx]
        
        # fit the model on the inner train set
        svc.fit(X_inner_train, y_inner_train)
        
        # evaluate the model on the inner test set
        inner_scores.append(svc.score(X_inner_test, y_inner_test))
    
    # calculate the mean inner score for this outer train-test split
    mean_inner_score = np.mean(inner_scores)
    
    # fit the model on the outer train set
    svc.fit(X_train, y_train)
    
    # evaluate the model on the outer test set
    loo_scores.append(svc.score(X_test, y_test))
```

```python
# calculate the mean score over all outer train-test splits
mean_score = np.mean(loo_scores)
print(mean_score)
```
